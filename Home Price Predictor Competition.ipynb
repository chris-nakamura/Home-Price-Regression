```python
import sys
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)
```

# Load Kaggle Data For Competition

```python
train_path = 'C:\\Users\\cnaka\\Documents\\house-prices-advanced-regression-techniques\\train.csv'
train_data_import = pd.read_csv(train_path)
train_data_import.head()
```

Identify Columns which need encoding

```python
string_columns_bool=train_data_import.dtypes=='object'
string_columns=train_data_import.columns[string_columns_bool]
num_columns = train_data_import.columns[string_columns_bool==False]
train_data_import[string_columns]
```

One-Hot Encode These Features

```python
train_data=pd.get_dummies(train_data_import)
train_data.head()
```

Find NA Values

```python
train_data.columns[train_data.isna().any()==True]
```

```python
#The Lot Frontage and Masonry Veneer Area values make sense to replace with 0 since they are measurements of area/length.  Garage Year Can ALso Be replaced with 0 since it is ordinal data
train_data=train_data.fillna(0)
```

Split Data

```python
train_x=train_data[train_data.columns[train_data.columns!="SalePrice"]]
train_y=train_data_import["SalePrice"]
```

```python
from sklearn import model_selection as ms
train_x,validate_x,train_y,validate_y=ms.train_test_split(train_x,train_y,train_size=.9)
```

# Basic Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor as DTR
```

```python
tree=DTR(max_depth=10,min_samples_leaf=10)
tree.fit(train_x,train_y)
```

```python
validation_results = pd.DataFrame([tree.predict(validate_x),validate_y]).transpose()
validation_results.columns=["Predicted","Actual"]
validation_results.head()
```

```python
RMSE=np.sqrt(np.mean((validation_results["Predicted"]-validation_results["Actual"])**2))
print("RMSE={}".format(RMSE))
```

```python
#Create function to replicate above
def Validation_Step(x,y,model):
    results = pd.DataFrame([model.predict(x),y]).transpose()
    results.columns=["Predicted","Actual"]
    RMSE=np.sqrt(np.mean((results["Predicted"]-results["Actual"])**2))
    print("RMSE={}".format(RMSE))
    return results
```

Identifying Important Features

```python
features=pd.DataFrame([tree.feature_names_in_,tree.feature_importances_]).transpose()
features.columns=["Feature","Importance"]
features=features.sort_values(by='Importance',ascending=False)
features
```

```python
features.loc[features["Importance"]>0,:]
```

# Dimension Reduction Using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=.99) #Preserver 99% of variance from original data
train_x_pca=pca.fit_transform(train_x)
pd.DataFrame(train_x_pca).head()
```

```python
tree_pca = DTR(max_depth=100,min_samples_leaf=10)
tree_pca.fit(train_x_pca,train_y)
```

```python
validate_x_pca=pca.transform(validate_x)
```

```python
Validation_Step(validate_x_pca,validate_y,tree_pca)
```

PCA has not provided a benefit to model generalization in this case

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor as RFR
rnd_frst = RFR(n_estimators = 500,min_samples_leaf = 10, max_depth = 100,n_jobs = -1)
rnd_frst.fit(train_x,train_y)
pred_v_act=Validation_Step(validate_x,validate_y,rnd_frst)
pred_v_act.head()
```

```python
rnd_frst_pca = RFR(n_estimators = 500,min_samples_leaf = 10, max_depth = 100,n_jobs = -1)
rnd_frst_pca.fit(train_x_pca,train_y)
pred_v_act=Validation_Step(validate_x_pca,validate_y,rnd_frst_pca)
pred_v_act.head()
```

# Run Analysis Excluding Dummy Variables

```python
num_train_data = train_data_import[num_columns]
num_train_data = num_train_data.fillna(0)
num_train_data.head()
```

```python
num_train_data_x = num_train_data[num_train_data.columns[num_train_data.columns!='SalePrice']]
num_train_data_y = num_train_data['SalePrice']
```

```python
num_train_data_x,num_validate_data_x,num_train_data_y,num_validate_data_y = ms.train_test_split(num_train_data_x,num_train_data_y,train_size=.9)
```

```python
rnd_frst_num = RFR(n_estimators = 500,min_samples_leaf = 10, max_depth = 100,n_jobs = -1)
rnd_frst_num.fit(num_train_data_x,num_train_data_y)
pred_v_act=Validation_Step(num_validate_data_x,num_validate_data_y,rnd_frst_num)
pred_v_act.head()
```

Excluding Dummy Variables has reduced overfitting.  It seems there are many datapoints which pare predicted well, but a few which are quite off.  We shall try a boosting technique to address this

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor as GBR
from sklearn.metrics import mean_squared_error as MSE
grad_boost = GBR(max_depth=100,min_samples_leaf = 10,n_estimators = 100,learning_rate = .1)
grad_boost.fit(num_train_data_x,num_train_data_y)
#Employ Early Stopping
errors = [MSE(num_validate_data_y,y_pred) for y_pred in grad_boost.staged_predict(num_validate_data_x)]
best_estimators = np.argmin(errors)+1
grad_best = GBR(max_depth=100,min_samples_leaf=10,n_estimators=best_estimators,learning_rate=.1)
grad_best.fit(num_train_data_x,num_train_data_y)

pred_v_act=Validation_Step(num_validate_data_x,num_validate_data_y,grad_best)
pred_v_act.head()
```

# Stacking Classifier with Ensemble of Random Forest and Gradient Boosted Models

## Here we will create a stack of models.  In the first layer we will have the gradient boost model and the random forest provide their estimates.  Then we will Train a classification model to differentiate which model will be best given the inputs.  The reasoning is that the random forest seems to do very well on some home types, but the gradien boosted model does better in most other cases.

```python
#Add both model predictions to validation set
stack_data = num_validate_data_x.merge(num_validate_data_y,left_index=True,right_index=True)
stack_data=stack_data.drop("Id",axis=1)
stack_data[["random_forest_estimate","gradient_boost_estimate"]] = pd.DataFrame(zip(rnd_frst_num.predict(num_train_data[num_train_data.columns[num_train_data.columns!='SalePrice']]),
    grad_best.predict(num_train_data[num_train_data.columns[num_train_data.columns!='SalePrice']])))
stack_data.head()
```

```python
#add identifier for if 
stack_data['better_model']=['A'  if abs(row['random_forest_estimate']-row['SalePrice'])<abs(row['gradient_boost_estimate']-row['SalePrice']) else 'B' for index,row in stack_data.iterrows()]
```

```python
stack_data_x = stack_data[stack_data.columns[[x not in ["better_model","random_forest_estimate","gradient_boost_estimate","SalePrice"] for x in stack_data.columns]]]
stack_data_y = stack_data["better_model"]
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
model_selector = RFC(n_estimators=100,min_samples_leaf = 3,max_depth=10,n_jobs = -1)
model_selector.fit(stack_data_x,stack_data_y)
```

```python
def final_prediction(x,model1,model2,model_selector):
    model1_pred = model1.predict(x)
    model2_pred = model2.predict(x)
    model_selection = model_selector.predict(x[x.columns[x.columns!='Id']])
    prediction = []
    for i in range(len(model_selection)):
        if model_selection[i]=='A':
            prediction.append(model1_pred[i])
        else:
            prediction.append(model2_pred[i])
    df = pd.DataFrame(zip(x["Id"],prediction))
    return df
```
